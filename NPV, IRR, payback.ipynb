```python
#Supose you have the follwoing cashflow
CFList=[-1000, 200, 200, 300, 500]
rate= 0.04

for n, cf in enumerate(CFList):
    print(n,cf)
```

```python
#calcule the NPV sugestion: iterate both by index and value using enumerate

NPV = 0

for n, cf in enumerate(CFList):
    NPV += cf/(1+rate)**(n)
print(NPV)
```

```python
#create a function called NPV that returns the NPV value

def npv(cfList,r):
    npv = 0
    for n, cf in enumerate(cfList):
        npv += cf/(1+r)**(n)
    return npv

npv(CFList,rate)
```

```python
#create a function called IRR that returns the IRR value

def irr(cfList):
    irr = 0
    for i in range (0,100):
        if npv(cfList,irr) > 0:
            irr += .001
            print(npv(CFList,irr))
            print(irr)
        else:
            irr -= .001
            print(npv(CFList,irr))
            print(irr)
    return irr

irr(CFList)
```
